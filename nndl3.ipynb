```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
temp=np.array([20,25,30,35,40]).reshape(-1,1)
sales=np.array([200,300,500,650,800])
```

```python
x=pd.DataFrame(temp)
y=pd.DataFrame(sales)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
predicted_sales=model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,predicted_sales)
```
